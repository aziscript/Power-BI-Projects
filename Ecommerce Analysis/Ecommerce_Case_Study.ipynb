---

#### üìù **Description**

In this Power BI case study, I analyzed sales and shipping data from **Whiskique**, a real-world ecommerce business specializing in pet supplies. The objective was to uncover sales performance insights, understand customer distribution, and build an executive dashboard to support decision-making across marketing, operations, and finance departments.

Using **Power Query**, I cleaned and prepared the data. With **DAX**, I built calculated measures to drive visualizations. The final result is an interactive, multi-page dashboard that highlights key business metrics and simulates the impact of different shipping scenarios.
