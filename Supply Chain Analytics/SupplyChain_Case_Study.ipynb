# 📦 Supply Chain Analytics – Make vs Buy Analysis with Power BI

## 🧠 Objective

To support Tenate Industries’ procurement team in making **data-driven "Make vs Buy" decisions** by building an **interactive Power BI model** that evaluates supplier quotes, internal manufacturing costs, capital investment, and quality implications across different order volumes.

## 🏭 Business Context

Tenate Industries frequently evaluates whether to **manufacture components internally** or **purchase them from suppliers**. This decision is influenced by:
- **Quoted supplier prices (Unit Cost + NRE)**
- **Internal manufacturing capabilities and capacity**
- **Capital investment requirements**
- **Volume-based cost changes**
- **Quality-related costs (based on yield rates)**

Traditional spreadsheets couldn’t efficiently handle this complexity. I was tasked with building a Power BI solution to enable dynamic scenario planning.

## 🧰 Tools & Skills Used
- **Power BI**
- **DAX (Data Analysis Expressions)**
- **Data modeling & relationships**
- **What-If parameter**
- **Row-Level Security**
- **Scenario planning and sensitivity analysis**
- **Cost analysis and supplier comparison**

## 💼 My Role

As a **Supply Chain Analyst**, I was responsible for:
- Collecting and preparing quote data
- Calculating full costs (Unit Cost × Volume + NRE)
- Modeling internal manufacturing costs and capacity
- Creating DAX-based scenario planners
- Building interactive dashboards
- Supporting strategic sourcing decisions
